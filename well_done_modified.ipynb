#### 1.2 Recognizing Variables in Dataset <a id = 3 ></a>

Variable definitions in the Dataset

- **_male_**: Gender of the participant (1 = male, 0 = female).
- **_age_**: Age of the participant.
- **_education_**: Education level of the participant.
- **_currentSmoker_**: Whether the participant is a current smoker (1 = yes, 0 = no).
- **_cigsPerDay_**: Number of cigarettes smoked per day by the participant.
- **_BPMeds_**: Whether the participant is on blood pressure medication (1 = yes, 0 = no).
- **_prevalentStroke_**: Whether the participant has had a prevalent stroke (1 = yes, 0 = no).
- **_prevalentHyp_**: Whether the participant has prevalent hypertension (1 = yes, 0 = no).
- **_diabetes_**: Whether the participant has diabetes (1 = yes, 0 = no).
- **_totChol_**: Total cholesterol level of the participant.
- **_sysBP_**: Systolic blood pressure of the participant.
- **_diaBP_**: Diastolic blood pressure of the participant.
- **_BMI_**: Body Mass Index of the participant.
- **_heartRate_**: Heart rate of the participant.
- **_glucose_**: Glucose level of the participant.
- **_TenYearCHD_**: Whether the participant developed coronary heart disease within ten years (1 = yes, 0 = no).

#### 3.1 Examining Missing Values <a id = 11 ></a>

```python
import missingno
missingno.bar(df, color = "b")
```

#### 3.1.1 Dealing With Missing Values <a id = 12 ></a>

```python
count=0
for i in df.isnull().sum(axis=1):
    if i>0:
        count=count+1
print('Total number of rows with missing values is ', count)
print('since it is only',round((count/len(df.index))*100), 'percent of the entire dataset the rows with missing values are excluded.')
```

```python
df.dropna(axis=0, inplace=True)
```

```python
isnull_number = []
for i in df.columns:
    x = df[i].isnull().sum()
    isnull_number.append(x)

pd.DataFrame(isnull_number, index = df.columns, columns = ["Total Missing Values"])
```

##### 3.5.1 Numerical Variable (Analysis with Histplot) <a id = 18></a>

```python
numeric_var
```

```python
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 10))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plotting each numeric variable
for i, var in enumerate(numeric_var):
    sns.histplot(df[var], kde=True, stat="density", ax=axes[i])
    axes[i].set_title(f'Distribution of {var}')
```

##### 3.5.1.1 Analysis Outputs(4) <a id = 19></a>

1. Age: Roughly normal distribution, centered around 45-50 years, ranging from about 30 to 70 years.

2. CigsPerDay: Right-skewed distribution with a large spike at 0 (non-smokers). Among smokers, there's a smaller peak around 20 cigarettes per day.

3. TotChol (Total Cholesterol): Approximately normal distribution, centered around 225-250 mg/dL, ranging from about 150 to 400 mg/dL.

4. SysBP (Systolic Blood Pressure): Slightly right-skewed, centered around 130-140 mmHg, ranging from about 90 to 250 mmHg.

5. DiaBP (Diastolic Blood Pressure): Approximately normal, centered around 80-85 mmHg, ranging from about 50 to 140 mmHg.

6. BMI: Slightly right-skewed, centered around 25-27, ranging from about 15 to 45.

7. HeartRate: Approximately normal, centered around 75-80 bpm, ranging from about 45 to 120 bpm.

8. Glucose: Right-skewed with a long tail, main peak around 80-90 mg/dL, ranging from about 50 to over 300 mg/dL.


#### 3.5.2 Categoriacal Variables (Analysis with Bar Chart)

```python
categoric_var
```

```python
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 10))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plotting each categorical variable
for i, var in enumerate(categoric_var):
    # Get the counts of each category
    data = df[var].value_counts()
    
    # Plot the bar chart with percentage labels
    axes[i].bar(data.index, data.values, label=data.index, color=plt.cm.tab20(range(len(data.index))))
    axes[i].bar_label(axes[i].containers[0], fmt=lambda x: f"{x/data.values.sum()*100:.1f}%")
    axes[i].set_title(f'Distribution of {var}')
    axes[i].set_ylabel('Counts')  # Set y-axis label
    
    # Set x-axis tick labels to individual categories
    axes[i].set_xticks(data.index)  # Set tick locations
    axes[i].set_xticklabels(data.index, rotation=45, ha='right')  # Set tick labels

# Adjust layout
plt.tight_layout()
plt.show()
```

##### Analysis Output

1. Male: 55.6% are female (0), 44.4% are male (1).

2. Education: Four levels, with 1.0 being the most common (41.7%), followed by 2.0 (30.1%), 3.0 (16.6%), and 4.0 (11.6%).

3. CurrentSmoker: 51.1% are non-smokers (0), 48.9% are current smokers (1).

4. BPMeds: 97.0% are not on blood pressure medication (0), 3.0% are on medication (1).

5. PrevalentStroke: 99.4% have no history of stroke (0), 0.6% have had a stroke (1).

6. PrevalentHyp: 68.8% do not have hypertension (0), 31.2% have hypertension (1).

7. Diabetes: 97.3% do not have diabetes (0), 2.7% have diabetes (1).

8. TenYearCHD: 84.8% did not develop coronary heart disease within 10 years (0), 15.2% did develop CHD (1).

Key observations:
- The dataset is fairly balanced in terms of gender and smoking status.
- Most participants have lower levels of education.
- A small portion of the population is on blood pressure medication.
- Stroke and diabetes are relatively rare in this population.
- About a third of the participants have hypertension.
- The target variable (TenYearCHD) shows that about 15% of participants developed coronary heart disease within 10 years, indicating an imbalanced classification problem.
